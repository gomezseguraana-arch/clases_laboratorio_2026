Codificación One-Hot en Aprendizaje Automático

La codificación one-hot es una técnica de preprocesamiento de datos que se utiliza para convertir variables categóricas en un formato numérico que los modelos de aprendizaje automático puedan comprender. Representa cada categoría como una columna binaria independiente, donde un valor de  1  indica la presencia de esa categoría y  0  indica su ausencia.

## Importancia de la codificación One-Hot

### Eliminación de la ordinalidad

Muchas variables categóricas no poseen un orden inherente, como por ejemplo “Masculino” y “Femenino”.
Si se asignaran valores numéricos directos (por ejemplo, Masculino = 0 y Femenino = 1), el modelo podría interpretar incorrectamente que existe una jerarquía u orden entre las categorías, generando predicciones sesgadas.

La codificación One-Hot evita este problema al representar cada categoría de manera independiente mediante columnas binarias.

### Mejora del rendimiento del modelo

La codificación One-Hot proporciona una representación más precisa de las variables categóricas, lo que puede mejorar el rendimiento de los modelos de aprendizaje automático.

Esto permite que los algoritmos detecten relaciones complejas dentro de los datos que podrían perderse si las categorías fueran tratadas como valores numéricos únicos.

### Compatibilidad con algoritmos

Muchos algoritmos de aprendizaje automático, especialmente aquellos basados en regresión lineal y descenso de gradiente, requieren datos numéricos como entrada.

La codificación One-Hot garantiza que las variables categóricas sean transformadas a un formato adecuado para su procesamiento por parte de estos modelos.


## Cómo funciona la codificación One-Hot

Para comprender mejor el concepto, veamos un ejemplo sencillo.
Imaginemos un conjunto de datos que contiene distintos tipos de frutas y sus precios. Inicialmente, cada fruta se representa mediante un valor categórico numérico.

### Datos originales

| Fruta   | Valor categórico | Precio |
| ------- | ---------------- | ------ |
| Manzana | 1                | 5      |
| Mango   | 2                | 10     |
| Manzana | 1                | 15     |
| Naranja | 3                | 20     |

El problema de esta representación es que el modelo podría interpretar incorrectamente que existe un orden entre las categorías, por ejemplo:

* Manzana = 1
* Mango = 2
* Naranja = 3

Esto podría hacer pensar al algoritmo que “Naranja” tiene un valor mayor o una relación matemática con “Manzana”, cuando en realidad solo son categorías distintas.

La codificación One-Hot soluciona este problema creando una columna independiente para cada categoría y utilizando valores binarios:

* 1 → indica presencia de la categoría
* 0 → indica ausencia de la categoría

### Resultado después de aplicar One-Hot Encoding

| Fruta   | Manzana | Mango | Naranja | Precio |
| ------- | ------- | ----- | ------- | ------ |
| Manzana | 1       | 0     | 0       | 5      |
| Mango   | 0       | 1     | 0       | 10     |
| Manzana | 1       | 0     | 0       | 15     |
| Naranja | 0       | 0     | 1       | 20     |

De esta manera, cada categoría queda representada de forma independiente, evitando interpretaciones incorrectas y permitiendo que los modelos de aprendizaje automático trabajen correctamente con variables categóricas.


## Implementación mediante Python

Para implementar la codificación One-Hot en Python, es posible utilizar bibliotecas como Pandas o Scikit-learn, ya que ambas ofrecen métodos simples y eficientes para transformar variables categóricas en variables numéricas binarias.

### Uso de Pandas

La biblioteca Pandas proporciona la función `get_dummies()`, que permite realizar codificación One-Hot de manera rápida y sencilla.

Este método transforma una variable categórica en múltiples columnas binarias, donde:

* `1` indica la presencia de la categoría.
* `0` indica la ausencia de la categoría.

Por ejemplo, una columna llamada `Gender` con valores `"M"` y `"F"` puede transformarse en columnas binarias como:

* `Gender_M`
* `Gender_F`

### Ejemplo en Python

```python
import pandas as pd

# Crear DataFrame
df = pd.DataFrame({
    'Gender': ['M', 'F', 'F', 'M']
})

# Aplicar One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Gender'])

print(df_encoded)
```

### Resultado

```python
   Gender_F  Gender_M
0         0         1
1         1         0
2         1         0
3         0         1
```

### Uso de `drop_first=True`

En algunos casos, es recomendable eliminar una de las columnas generadas para evitar problemas de multicolinealidad en modelos estadísticos y de regresión lineal.

Esto se puede hacer utilizando el parámetro:

```python
drop_first=True
```

### Ejemplo

```python
df_encoded = pd.get_dummies(df, columns=['Gender'], drop_first=True)

print(df_encoded)
```

### Resultado

```python
   Gender_M
0         1
1         0
2         0
3         1
```

En este caso, la columna `Gender_F` se elimina y queda únicamente `Gender_M`.
Cuando el valor es:

* `1` → corresponde a `"M"`
* `0` → corresponde a `"F"`

Esto reduce redundancia y evita relaciones lineales innecesarias entre variables.


````python
import pandas as pd

# Crear DataFrame
df = pd.DataFrame({
    'Gender': ['M', 'F', 'F', 'M']
})
print(df)
````

````python

# Aplicar One-Hot Encoding
df_encoded = pd.get_dummies(df, columns=['Gender'])
df_encoded = pd.get_dummies(df, columns=['Gender'], dtype=int)


print(df_encoded)
````

````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {
    'Employee id': [10, 20, 15, 25, 30],
    'Gender': ['M', 'F', 'F', 'M', 'F'],
    'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice']
}

df = pd.DataFrame(data)

print(f"Datos Originales:\n{df}\n")

# Columnas categóricas
categorical_columns = ['Gender', 'Remarks']

# One-Hot Encoding con pandas
df_pandas_encoded = pd.get_dummies(
    df,
    columns=categorical_columns,
    drop_first=True,
    dtype=int
)

print(f"One-Hot Encoding con Pandas:\n{df_pandas_encoded}\n")

# One-Hot Encoding con Scikit-Learn
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[categorical_columns])

one_hot_df = pd.DataFrame(
    one_hot_encoded,
    columns=encoder.get_feature_names_out(categorical_columns)
)

# Combinar columnas originales numéricas con las codificadas
df_sklearn_encoded = pd.concat(
    [df.drop(categorical_columns, axis=1), one_hot_df],
    axis=1
)

print(f"One-Hot Encoding con Scikit-Learn:\n{df_sklearn_encoded}\n")
````

Podemos observar que tenemos 3 columnas de Observaciones y 2 de Género en los datos. Sin embargo, se pueden usar n-1 columnas para definir parámetros si tienen n etiquetas únicas. Por ejemplo, si solo conservamos la columna Género_Femenino y eliminamos la columna Género_Masculino , podemos transmitir toda la información: cuando la etiqueta es 1, significa mujer, y cuando la etiqueta es 0, significa hombre. De esta manera, podemos codificar los datos categóricos y reducir el número de parámetros.

## Uso de la biblioteca Scikit-Learn

Scikit-learn (`sklearn`) es una de las bibliotecas de aprendizaje automático más utilizadas en Python y proporciona múltiples herramientas para el preprocesamiento de datos.

Entre ellas se encuentra `OneHotEncoder`, una función utilizada para transformar variables categóricas en vectores binarios mediante codificación One-Hot.

### Selección de columnas categóricas

Antes de aplicar One-Hot Encoding, es necesario identificar qué columnas contienen datos categóricos.

Esto puede hacerse utilizando:

```python id="v9d2qp"
df.select_dtypes(include=['object'])
```

### ¿Qué hace este método?

El método:

```python id="e1f7kx"
select_dtypes(include=['object'])
```

selecciona únicamente las columnas cuyo tipo de dato es `object`, que normalmente corresponde a texto o variables categóricas.

Por ejemplo, en un DataFrame como el siguiente:

```python id="m4q8sn"
   Employee id Gender Remarks
0           10      M    Good
1           20      F    Nice
2           15      F    Good
```

las columnas categóricas detectadas serían:

```python id="q3w5lt"
['Gender', 'Remarks']
```



````python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

data = {'Employee id': [10, 20, 15, 25, 30],
        'Gender': ['M', 'F', 'F', 'M', 'F'],
        'Remarks': ['Good', 'Nice', 'Good', 'Great', 'Nice'],
        }
df = pd.DataFrame(data)
print(f"Employee data : \n{df}")

categorical_columns = df.select_dtypes(include=['object']).columns.tolist()
encoder = OneHotEncoder(sparse_output=False)

one_hot_encoded = encoder.fit_transform(df[categorical_columns])

one_hot_df = pd.DataFrame(one_hot_encoded, columns=encoder.get_feature_names_out(categorical_columns))

df_encoded = pd.concat([df, one_hot_df], axis=1)

df_encoded = df_encoded.drop(categorical_columns, axis=1)
print(f"Encoded Employee data : \n{df_encoded}")
````

## Alternativas a la codificación One-Hot

Si bien la codificación One-Hot es una de las técnicas más utilizadas para trabajar con variables categóricas, existen otras alternativas que pueden resultar más adecuadas dependiendo del tipo de datos y del problema de aprendizaje automático.

### Codificación de etiquetas (Label Encoding)

La codificación de etiquetas consiste en asignar un número entero único a cada categoría.

Por ejemplo:

* Bajo → 0
* Medio → 1
* Alto → 2

Este método es útil cuando las categorías poseen un orden natural u ordinal.
Sin embargo, no es recomendable para variables nominales, ya que el modelo podría interpretar incorrectamente que existe una relación jerárquica entre los valores.

### Codificación binaria (Binary Encoding)

La codificación binaria combina características de la codificación de etiquetas y de la codificación One-Hot.

Primero convierte cada categoría en un número entero y luego transforma esos números en representación binaria.

Esta técnica permite:

* Reducir la cantidad de columnas generadas.
* Disminuir la dimensionalidad del conjunto de datos.
* Mantener información relevante de las categorías.

Es especialmente útil cuando existen muchas categorías diferentes.

### Codificación de destino (Target Encoding)

La codificación de destino reemplaza cada categoría por una estadística calculada a partir de la variable objetivo, generalmente la media.

Por ejemplo, si una categoría tiene mayor promedio de aprobación o ventas, ese valor promedio reemplaza a la categoría original.

Este método puede ser muy efectivo cuando existen variables categóricas con una gran cantidad de valores únicos.

Sin embargo, debe utilizarse cuidadosamente, ya que puede provocar fuga de información (*data leakage*) si se aplica incorrectamente durante el entrenamiento del modelo.


## Ejemplo de Codificación Binaria (Binary Encoding)

La codificación binaria transforma las categorías en números enteros y luego convierte esos números a representación binaria.

Para este tipo de codificación puede utilizarse la biblioteca `category_encoders`.

### Posible resultado


En este caso:

* Cada categoría se transforma primero en un número.
* Luego ese número se convierte a formato binario.
* Esto reduce la cantidad de columnas en comparación con One-Hot Encoding.

## Ejemplo de Codificación de Destino (Target Encoding)

La codificación de destino reemplaza cada categoría por la media de la variable objetivo asociada a esa categoría.

### Código en Python

```python id="y3m9qa"
import pandas as pd
import category_encoders as ce

# Datos de ejemplo
df = pd.DataFrame({
    'Ciudad': ['Catamarca', 'Córdoba', 'Catamarca', 'Salta', 'Córdoba'],
    'Ventas': [100, 200, 150, 120, 250]
})

print("Datos originales:")
print(df)

# Crear codificador Target Encoding
encoder = ce.TargetEncoder(cols=['Ciudad'])

# Aplicar codificación usando la variable objetivo
df['Ciudad_encoded'] = encoder.fit_transform(
    df['Ciudad'],
    df['Ventas']
)

print("\nTarget Encoding:")
print(df)
```

### Posible resultado

```python id="u8k1vf"
      Ciudad  Ventas  Ciudad_encoded
0  Catamarca      100            125
1    Córdoba      200            225
2  Catamarca      150            125
3      Salta      120            120
4    Córdoba      250            225
```

### Explicación

* `Catamarca` → promedio de ventas = `(100 + 150) / 2 = 125`
* `Córdoba` → promedio de ventas = `(200 + 250) / 2 = 225`
* `Salta` → promedio de ventas = `120`

Cada categoría es reemplazada por el promedio de la variable objetivo correspondiente.
